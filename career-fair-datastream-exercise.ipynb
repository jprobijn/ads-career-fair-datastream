### The datastream

We've gone ahead and created an example regular datastream which describes the positions of player positions in a football match over the course of 1 minute. Before we start doing some fun things with this datastream, let's first take a quick look at what kind of structure the incoming data will have.

```python
datastream.get_data_sample()
```

For those of you familiar with JSON, this should be easy enough to understand. You will get a message like this for every timepoint in the datastream. At the top level there's an object with 2 fields, *Timestamp* and *pos*.

- *Timestamp* will contain a datetime string describing the exact time the measurements took place.
- *pos* will contain an array of objects where each object has 4 fields describing a particular participant (player or ball) at that moment

Every participant is described by an object containing several fields:

- *PlrID* contains the unique identifier for that particular player, which is not going to change at any point in the datastream
- *TeamName* contains a string describing the player's team. For the ball this will always be *None*
- *X* contains the X-coordinate of the participant (position across the length of the field)
- *Y* contains the Y-coordinate of the participant. (position across the width of the field)

A little note on the coordinates: The center of the field is (0,0) and the X coordinates can vary from -52.5 to 52.5 while the Y coordinates can vary from -34 to 34.
